# Empirical time-complexity

```python
from sky_inv_quiv import computeHN
import sky_inv_quiv.auxHN as aux
from sky_inv_quiv import Field
from sky_inv_quiv.quiver_parser import parse_quiver_file
from time import time
```

### Over $\mathbb Q$

```python
for field_str in ["Q","F_2"]:
    print(f"Field: {field_str}")
    for (ex_name,nmax) in [(f"torus_dim{d}_1_",5) for d in range(2,5)]+[("two_circles_",5)]:
        
        print(f"Example: {ex_name}")
        f = open(f"output_{ex_name}{field_str}.txt", "w")
        for n in range(2,6):
            print(n)
            t0 = time()
            print(n,file=f)

            Q = parse_quiver_file(f"{ex_name}{n}x{n}.txt", field=Field(field_str), grid_flag=True)
            HN = computeHN(Q, [(0,0)])[(0,0)]
            for t in HN:
                print(t[0], end=" ", file=f)
            print("",file=f)
            print(time() - t0,file=f)
        f.close()
```
